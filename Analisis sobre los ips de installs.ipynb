```python
import numpy as np 
import pandas as pd

# plots
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline

installs = pd.read_csv('data/installs.csv',low_memory=False)
```

```python
color = cm.inferno_r(np.linspace(.4,.8, 10))
plot = installs['ip_address'].value_counts().head(15).plot(kind="barh",color = color,figsize=(12,10))
plot.set_ylabel('Ips',fontsize=16)
plot.set_xlabel('Cantidad de instalaciones',fontsize=16)
plot.set_title('Ips desde donde se instalan aplicaciones', fontsize=20)
plt.show()
```
