```python
import pandas as pd
import numpy as np

# plots
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
%matplotlib inline
from wordcloud import WordCloud ## poner en consola 'python -m pip install wordcloud'
```

```python
installs = pd.read_csv("../data/installs.csv",low_memory= False);
```

```python
installs.info()
```

```python
installs.columns
```

```python
listaDeTerminosBuscados = installs['kind'].value_counts().reset_index()['index'].tolist()
listaDeTerminosBuscados
```

```python
k = []
for i in listaDeTerminosBuscados:
    j = i.replace(' ','')
    j = j.lower()
    k.append(j)
    
listaDeTerminosBuscados = k
text = ' '.join(listaDeTerminosBuscados)
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=1480, height=1480, background_color="green" ,margin=0,colormap="Reds", max_words=250).generate(text)
plt.figure(figsize=[14,19]) 
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()
```

##### Como se puede observar, las instalaciones realizadas en Jammps se centran en muchas ocurrencias que estan vinculadas con OPEN ( ya sea desde eventos para abrir la app) y LOGUEOS en la página.
